```python
import numpy as np
from classy import Class

# initialize class
cosmo = Class()

# Planck 2018 parameters from https://arxiv.org/pdf/1807.06209.pdf
parameters = {
    "omega_b": 0.02233,
    "omega_cdm": 0.1198,
    "h": 0.6737,
    "A_s": 2.097e-9,
    "n_s": 0.9652,
    "tau_reio": 0.0540,
}

# set the class parameters
cosmo.set({"output": "mPk", "P_k_max_1/Mpc": 10.0, **parameters})

# run class
cosmo.compute()

# set up an array for k
k_min = 0.1
k_max = 5
n_points = 100

K = np.logspace(np.log(k_min), np.log10(k_max), n_points)

# calculate Pk at all k values (note the normalization for k and P_k
h = parameters["h"]
P_k = [cosmo.pk(k * h, 0) * h ** 3 for k in K]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data = np.loadtxt("../data/lyman-alpha.txt")
data[:, 0] /= h  # renormalize k, because usually units are h / Mpc

fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.loglog(K, P_k)
ax.errorbar(*data.T, marker="o", ms=5, ls="none")
ax.set_xlabel("k [1/Mpc]")
ax.set_ylabel("p(k)")
```
